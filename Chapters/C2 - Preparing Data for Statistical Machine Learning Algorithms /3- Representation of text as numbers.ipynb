## **Byte level tokenization**

````python
def byte_level_tokenize(text):
    return [c for c in text.encode('utf-8')]

text = "Tokenization is essential. 😄"
byte_tokens = byte_level_tokenize(text)
print(byte_tokens)
````

**Explanation**  


1. **Function Definition**:
   - `byte_level_tokenize(text)`: A function that converts the input string (`text`) into its byte representation using UTF-8 encoding.

2. **How It Works**:
   - `text.encode('utf-8')`: Encodes the input string into a sequence of bytes.
   - The list comprehension `[c for c in ...]` iterates over these bytes and stores them as individual integers.

3. **Example Input**:
   - The string `"Tokenization is essential. 😄"`.

4. **Function Output**:
   - The output is a list of integers, each representing a byte in the UTF-8 encoding of the string, including spaces, punctuation, and emojis.

5. **Output Display**:
   - The result of byte-level tokenization is stored in `byte_tokens` and printed.

## **Word level tokenization**

````python
import nltk
nltk.download('punkt_tab')
from nltk.tokenize import word_tokenize
text = "Tokenization is essential. 😄"

def word_tokenize_nltk(text):
    return word_tokenize(text)

word_tokens_nltk = word_tokenize_nltk(text)
print(f"NLTK word level tokenization {word_tokens_nltk}")

from spacy.lang.en import English

nlp = English()
def word_tokenize_spacy(text):
    return [token.text for token in nlp(text)]

word_tokens_spacy = word_tokenize_spacy(text)
print(f"Spacy word level tokenization {word_tokens_spacy}")

import re

def word_tokenize_regex(text):
    return re.findall(r'\b\w+\b', text)

word_tokens_regex = word_tokenize_regex(text)
print(f"Regex word level tokenization {word_tokens_regex}")
````

**Explanation**

This code demonstrates three different methods for tokenizing text into individual words: NLTK, spaCy, and regular expressions. Here’s a detailed breakdown:


1. **NLTK Word Tokenization**:
   - Uses `nltk.tokenize.word_tokenize` to split the text into tokens.
   - Handles punctuation and emojis effectively as separate tokens.
   - Requires downloading the `punkt` tokenizer with `nltk.download('punkt')`.


2. **spaCy Word Tokenization**:
- Utilizes spaCy’s `English` tokenizer to split the text.
- Handles punctuation and emojis like NLTK, treating them as separate tokens.
- Tokens are extracted using a list comprehension: `[token.text for token in nlp(text)]`.



3. **Regex Word Tokenization**:
- Uses `re.findall` with the pattern `\b\w+\b` to extract word-like tokens.
- Matches word characters (`\w`) surrounded by word boundaries (`\b`).
- Does not handle emojis or punctuation as separate tokens.


## **Sub-word level tokenization: Byte pair encoding**

````python
import re
from collections import defaultdict

def get_stats(vocab):
    """
    Given a vocabulary (dictionary mapping words to frequency counts), returns a
    dictionary of tuples representing the frequency count of pairs of characters
    in the vocabulary.
    """
    pairs = defaultdict(int)
    for word, freq in vocab.items():
        symbols = word.split()
        for i in range(len(symbols)-1):
            pairs[symbols[i],symbols[i+1]] += freq
    return pairs

def merge_vocab(pair, v_in):
    """
    Given a pair of characters and a vocabulary, returns a new vocabulary with the
    pair of characters merged together wherever they appear.
    """
    v_out = {}
    bigram = re.escape(' '.join(pair))
    p = re.compile(r'(?<!\S)' + bigram + r'(?!\S)')
    for word in v_in:
        w_out = p.sub(''.join(pair), word)
        v_out[w_out] = v_in[word]
    return v_out

def get_vocab(data):
    """
    Given a list of strings, returns a dictionary of words mapping to their frequency
    count in the data.
    """
    vocab = defaultdict(int)
    for line in data:
        for word in line.split():
            vocab[' '.join(list(word)) + ' </w>'] += 1
    return vocab

def byte_pair_encoding(data, n):
    """
    Given a list of strings and an integer n, returns a list of n merged pairs
    of characters found in the vocabulary of the input data.
    """
    vocab = get_vocab(data)
    for i in range(n):
        pairs = get_stats(vocab)
        best = max(pairs, key=pairs.get)
        vocab = merge_vocab(best, vocab)
    return vocab

# Example usage:
corpus = '''low low low.
lower lower lower.
new new new.
newer newer newer.'''
data = corpus.split('.')

n=6
bpe_pairs = byte_pair_encoding(data, n)
bpe_pairs
````

**Explanation**

This code implements **Byte Pair Encoding (BPE)**, a subword tokenization algorithm widely used in NLP for compressing text and creating subword units. Here's a detailed breakdown:


1. **`get_vocab(data)`**:
   - Converts a list of strings into a vocabulary with word frequencies.  
   - Each word is split into individual characters, with `</w>` appended to indicate the end of the word.  
   - Example: `"low"` becomes `"l o w </w>"`.

2. **`get_stats(vocab)`**:
   - Computes the frequency of adjacent character pairs (bigrams) in the vocabulary.  
   - Returns a dictionary where keys are character pairs, and values are their frequency.

3. **`merge_vocab(pair, v_in)`**:
   - Merges the most frequent pair of characters into a single unit.  
   - Updates the vocabulary by replacing occurrences of the pair with their merged form.

4. **`byte_pair_encoding(data, n)`**:
   - Applies the BPE algorithm for `n` iterations:
     - Creates an initial vocabulary using `get_vocab`.  
     - Repeatedly finds and merges the most frequent character pair using `get_stats` and `merge_vocab`.  
   - Returns the updated vocabulary after `n` merges.

## **Word peiece tokenization**

````python
from transformers import BertTokenizer

def wordpiece_tokenization(text):
    # Initialize the tokenizer with a pre-trained BERT model
    tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')

    # Tokenize the text
    tokens = tokenizer.tokenize(text)

    return tokens
````

````python
sample_text = "Tokenization helps in handling rare words effectively."
tokens = wordpiece_tokenization(sample_text)

print("Tokens:", tokens)
````

**Explanation**

This code demonstrates the **WordPiece tokenization** method using a pre-trained BERT tokenizer from the Hugging Face `transformers` library. Here’s a detailed breakdown:

---

#### Key Steps

1. **Initialize the BERT Tokenizer**:
   - The tokenizer is loaded using `BertTokenizer.from_pretrained('bert-base-uncased')`.  
   - `bert-base-uncased` is a pre-trained BERT model that performs case-insensitive tokenization.

2. **Tokenize Input Text**:
   - The `tokenizer.tokenize(text)` method splits the input sentence into subword tokens based on the WordPiece algorithm.  
   - Common words remain as whole tokens (e.g., `"helps"`), while rare or unknown words are broken into subword units prefixed by `"##"` (e.g., `"handling"` becomes `["hand", "##ling"]`).

3. **Output Tokens**:
   - The resulting tokens are printed as a list, showing how the input text has been tokenized.

- `##` indicates that the subword is part of a larger word (e.g., `"##ization"` continues `"token"`).  
- Words like `"helps"` remain intact because they are present in the tokenizer's vocabulary.

---

#### Applications

1. **Subword Tokenization**:
   - Efficiently handles rare or out-of-vocabulary words by splitting them into meaningful subword units.
2. **Preprocessing for BERT**:
   - Prepares text for input to BERT models, ensuring compatibility with its subword-based vocabulary.
3. **Improved Generalization**:
   - Reduces vocabulary size while retaining the ability to represent unseen words.

#### Advantages of WordPiece Tokenization

## **Bag of words**

````python
from collections import Counter

def bag_of_words(text):
    # Tokenize the text (you can use any tokenization method you prefer)
    tokens = word_tokenize_regex(text.lower())  # Convert to lowercase for consistency

    # Remove punctuation and other non-alphanumeric characters
    tokens = [token for token in tokens if token.isalnum()]

    # Count the frequency of each word
    word_counts = Counter(tokens)

    return word_counts

# Example usage
text = "This is a sample text. This text is used for bag of words demonstration."
word_counts = bag_of_words(text)
word_counts
````

**Explanation**

This code implements a **Bag of Words (BoW)** model to represent text data as a collection of word frequencies. Here's a detailed breakdown:

---

#### Key Steps

1. **Tokenize Text**:
   - Uses a simple tokenization method (`word_tokenize_regex`) to split the input text into words.  
   - Converts the text to lowercase (`text.lower()`) to ensure case-insensitive tokenization.

2. **Filter Tokens**:
   - Removes punctuation and non-alphanumeric characters using `isalnum()`.  
   - Ensures that only valid words are retained for analysis.

3. **Count Word Frequencies**:
   - Uses Python's `Counter` from the `collections` module to count the occurrence of each word in the tokenized text.

4. **Return Word Counts**:
   - The result is a dictionary-like `Counter` object, where keys are words and values are their frequencies.

- Words are normalized to lowercase, and their frequencies are displayed.

---

#### Applications

1. **Text Representation**:
   - Transforms text into numerical data for machine learning models.
2. **Feature Extraction**:
   - Extracts simple features (word counts) for tasks like text classification, sentiment analysis, and topic modeling.
3. **Preprocessing**:
   - Forms the basis for advanced techniques like Term Frequency-Inverse Document Frequency (TF-IDF).

---

#### Advantages of BoW

1. **Simplicity**:
   - Easy to implement and understand.
2. **Efficiency**:
   - Works well for small to medium-sized datasets.

#### Limitations of BoW

1. **Loss of Context**:
   - Ignores word order and meaning.
2. **Dimensionality**:
   - For large vocabularies, the feature space becomes sparse and high-dimensional.



## **TF-IDF**

image.png

````python
import math
from collections import Counter
import re

def word_tokenize_regex(text):
    # Simple regex-based tokenizer
    return re.findall(r'\w+', text)

def calculate_term_frequency(token, tokens_in_document):
    # Calculate term frequency for a specific token in the document
    frequency_of_all_tokens = Counter(tokens_in_document)
    frequency_of_token = frequency_of_all_tokens[token]
    return frequency_of_token / len(tokens_in_document)

def calculate_inverse_document_frequency(token, corpus):
    # Calculate IDF for a specific token
    number_of_documents_that_contain_token = sum(1 for document in corpus if token in word_tokenize_regex(document.lower()))
    number_of_documents_in_corpus = len(corpus)
    return math.log(number_of_documents_in_corpus / (1 + number_of_documents_that_contain_token))

def tfidf(document, corpus):
    tokens_in_document = word_tokenize_regex(document.lower())
    tfidf_scores = {}
    for token in tokens_in_document:
        term_frequency = calculate_term_frequency(token, tokens_in_document)
        inverse_document_frequency = calculate_inverse_document_frequency(token, corpus)
        tfidf_scores[token] = term_frequency * inverse_document_frequency
    return tfidf_scores

# Example corpus
corpus = [
    'This is how TFIDF works.',
    'It is a vectorization technique.',
    'It can be used to understand the properties of a text'
]

# Calculate TF-IDF for each document
for idx, document in enumerate(corpus):
    print(f"Document {idx + 1}: {document}")
    tfidf_scores = tfidf(document, corpus)
    for token, score in tfidf_scores.items():
        print(f"  Token: {token}, TF-IDF Score: {score:.4f}")
````

**Explanation**

This code implements **TF-IDF (Term Frequency-Inverse Document Frequency)**, a technique for evaluating the importance of terms in a document relative to a corpus. Here's a detailed explanation:

---

#### Key Steps

1. **Tokenization**:
   - `word_tokenize_regex(text)` splits text into words using a simple regex pattern (`\w+`), ensuring only alphanumeric characters are extracted.

2. **Calculate Term Frequency (TF)**:
   - `calculate_term_frequency(token, tokens_in_document)` computes the term frequency of a token as:
     \[
     TF = \frac{\text{Frequency of the token in the document}}{\text{Total number of tokens in the document}}
     \]

3. **Calculate Inverse Document Frequency (IDF)**:
   - `calculate_inverse_document_frequency(token, corpus)` calculates the IDF as:
     \[
     IDF = \log{\frac{N}{1 + \text{n}}}
     \]
     Where:
     - \( N \) is the total number of documents in the corpus.
     - \( \text{n} \) is the number of documents containing the token.
     - Adding 1 in the denominator avoids division by zero.

4. **Compute TF-IDF**:
   - `tfidf(document, corpus)` calculates the TF-IDF for each token in a document as:
     \[
     TF\text{-}IDF = TF \times IDF
     \]
   - Scores are stored in the dictionary `tfidf_scores`.

5. **Output Results**:
   - For each document, the code prints the TF-IDF scores of its tokens.


---

#### Applications

1. **Text Analysis**:
   - Identifies the most important terms in a document relative to the entire corpus.
2. **Feature Extraction**:
   - Converts text data into numerical vectors for machine learning models.
3. **Information Retrieval**:
   - Helps search engines rank documents by relevance to a query.

---

#### Advantages of TF-IDF

1. **Contextual Importance**:
   - Highlights terms that are unique to a document while downweighting common terms.
2. **Simplicity**:
   - Easy to compute and interpret.
3. **Effective for Small Datasets**:
   - Performs well on small to medium-sized text corpora.

---

#### Summary

This code computes the TF-IDF scores for tokens in each document of a given corpus. It highlights the terms that are most relevant to each document, enabling effective text analysis and preprocessing for NLP tasks.

# TF-IDF Implementation

### Key Steps
1. **Tokenization**: Splits text into words using `\w+` regex.
2. **Term Frequency (TF)**:
   - Measures how often a token appears in a document.
   - Formula: `TF = (token count) / (total tokens)`.
3. **Inverse Document Frequency (IDF)**:
   - Measures how unique a token is across the corpus.
   - Formula: `IDF = log(total docs / (1 + docs with token))`.
4. **TF-IDF**:
   - Combines TF and IDF to score token importance.
   - Formula: `TF-IDF = TF * IDF`.

### Example
- **Corpus**: 3 documents about TF-IDF.
- **Output**: Displays tokens and their TF-IDF scores for each document.

### Why TF-IDF?
- Highlights important words.
- Reduces noise from common terms.
